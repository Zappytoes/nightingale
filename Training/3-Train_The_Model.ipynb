# <span style='color:Blue'> NOTEBOOK 3: Train the SCRDet++ model using the Nightingale framework

## If you are making no other changes to your configuration file (libs/configs/cfgs.py) and you have completed the first two notebooks (1-Nightingale_Format_Groundtruth.ipynb & 2-Make_TF_Record.ipynb), then you are ready to train the model on the Tensorflow Record you created in Notebook #2. There is no code to run in this notebook, rather, here there are instructions for running a few simple commands in the terminal. 

## <span style='color:red'> Note for tutorial: If you are training the model using only the small number of images and annotations provided for the tutorial with the Git Repo, change the SAVE_WEIGHTS_INTE variable in your cfgs.py file to 
    
>>> ### `SAVE_WEIGHTS_INTE = 1000`
    
## <span style='color:red'> The SAVE_WEIGHTS_INTE controls the weights checkpointing frequency and the total number of training iterations. Setting SAVE_WEIGHTS_INTE = 1000 will cause training to end more quickly for purposes of this tutorial. For training the model with a complete dataset (i.e, hundreds or thousands of images and annotations) it is recomended that SAVE_WEIGHTS_INTE be set to a value in the range of tens of thousands (e.g., 10000, 20000, 50000 etc), but training iterations are another hyperparameter to play with. 


# <span style='color:Blue'> To train the model, run multi_gpu_train.py (tools/multi_gpu_train.py) from the terminal

##  Run "multi_gpu_train.py" from the terminal:

> ### `conda activate nightingale_env`

> ### `cd Nightingale/Training/tools`

> ### `python multi_gpu_train.py`

### <span style='color:magenta'> Note that multi_gpu_train will pickup at the last saved weight (in output/trained_weights) if training is shut down for any reason. 



## After several minutes, you will start to see output to the terminal reporting training progress, e.g.,

```
2021-11-04 14:39:34: global_step:200  current_step:200
per_cost_time:12.412s
rpn_cls_loss:0.624
rpn_reg_loss:0.014
fast_cls_loss:0.144
fast_reg_loss:0.000
mask_loss:0.074
total_losses:0.856
```
## Your training will run for MAX_ITERATION times (see your config file). You can let training complete, or, once some weights have been saved in your output/trained_weights folder, you can end training with Ctrl+c. <span style='color:red'> You can stop training at any time with Ctrl+c


## Once you run "multi_gpu_train.py", a new folder called "output" will appear in the Training directory. As training progresses, the "output" folder will eventually contain two subdirectories:
* ## summary
> ### -> The event file(s) for tracking progress in tensorboard will be stored here
* ## trained_weights
> ### -> The tensorflow checkpoint files will be stored in here

# <span style='color:Blue'> Once the model is training, you will be able to view progress in Tensorboard. Open a new terminal and do:

> ### `conda activate nightingale_env`

> ### `cd Nightingale/Training/ouput/summary`

## Depending on your system configuration, you may need to specify a hostname and port for Tensorboard to run on. You can try the following commands and see which one works:
> ### `tensorboard --logdir=.` 
> ### `tensorboard --logdir . --host ${HOSTNAME}.east.ile.nga.ic.gov --port 7777`
